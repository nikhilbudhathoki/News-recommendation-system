### MULTI LANGUAGE STEMMMING

```python
import re
import nltk
from nltk.stem import PorterStemmer

# Initialize the English stemmer
ps = PorterStemmer()

# List of common Nepali suffixes (can be expanded)
nepali_suffixes = ["हरू", "मा", "ले", "लाई", "बाट", "को", "का", "की", "सँग", "हरूलाई", "हरु", "ति", "ता", "ए", "ने"]

def is_nepali(word):
    """Check if a word is likely Nepali based on Unicode range."""
    return any("\u0900" <= char <= "\u097F" for char in word)

def custom_nepali_stem(word):
    """Manually remove common Nepali suffixes."""
    for suffix in nepali_suffixes:
        if word.endswith(suffix):
            return re.sub(suffix + "$", "", word)  # Remove suffix
    return word  # Return word if no suffix matched

def bilingual_stemmer(text):
    words = text.split()
    stemmed_words = []

    for word in words:
        clean_word = re.sub(r"[^a-zA-Z\u0900-\u097F]", "", word)  # Remove special characters

        if is_nepali(clean_word):
            stemmed_word = custom_nepali_stem(clean_word)  # Apply Nepali stemming
        else:
            stemmed_word = ps.stem(clean_word)  # Apply English stemming
        
        stemmed_words.append(stemmed_word)

    return " ".join(stemmed_words)


```

### Uses TF-IDF (Term Frequency-Inverse Document Frequency) to convert news text into numerical vectors
Considers both single words and pairs of words (ngram_range=(1, 2))
Removes common English stop words
Limits to 10,000 most important features
Calculates similarity between articles using cosine similarity

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import scipy.sparse as sp

class NewsRecommender:
    def __init__(self, data_path):
        """Initialize the recommender system with data path"""
        self.cleaned_data = pd.read_csv(data_path)
        self.prepare_data()
        self.create_similarity_matrices()
        
    def prepare_data(self):
        """Prepare and validate the data"""
        # Aggregate news content
        self.news_grouped = self.cleaned_data.groupby('News ID')['Combined'].agg(
            lambda texts: ' '.join(texts.astype(str))
        ).reset_index()
        
        # Add Main Category
        self.news_grouped = self.news_grouped.merge(
            self.cleaned_data[['News ID', 'Main Category']].drop_duplicates(),
            on='News ID',
            how='left'
        )
        
        # Create news ID mapping
        self.news_ids = self.news_grouped['News ID'].tolist()
        self.news_to_idx = {news_id: idx for idx, news_id in enumerate(self.news_ids)}
        
    def create_similarity_matrices(self):
        """Create content and category similarity matrices"""
        # Content-based similarity
        vectorizer = TfidfVectorizer(
            stop_words='english',
            #remove stopwords again according to the language
            ngram_range=(1, 2),
            max_df=0.85,
            min_df=2,
            sublinear_tf=True,
            max_features=10000
        )
        news_embeddings = vectorizer.fit_transform(self.news_grouped['Combined'])
        self.content_similarity_matrix = cosine_similarity(news_embeddings, dense_output=False)
        
        
        categories = self.news_grouped['Main Category'].unique()
        category_to_idx = {cat: idx for idx, cat in enumerate(categories)}
        category_matrix = np.zeros((len(self.news_grouped), len(categories)))
        
        for idx, row in self.news_grouped.iterrows():
            category_matrix[idx, category_to_idx[row['Main Category']]] = 1
            
        self.category_similarity_matrix = cosine_similarity(category_matrix, dense_output=False)
        
        # Create hybrid similarity matrix
        self.hybrid_similarity_matrix = 0.6 * self.content_similarity_matrix + 0.4 * self.category_similarity_matrix
        
    def get_user_history(self, user_id):
        """Get the news reading history for a user"""
        return self.cleaned_data[self.cleaned_data['User ID'] == user_id]['News ID'].unique()
        
    def get_news_details(self, news_id):
        """Get details of a specific news article"""
        news_info = self.cleaned_data[self.cleaned_data['News ID'] == news_id].iloc[0]
        return {
            'news_id': news_id,
            'title': news_info['News Title'],
            'content': news_info['News'],
            'category': news_info['Main Category']
        }
        
    def recommend_news(self, user_id, top_n=5):
        """Generate news recommendations for a user"""
        # Get user's reading history
        user_clicked = self.get_user_history(user_id)
        
        if not user_clicked.size:
            print(f"User {user_id} has no click history.")
            return []
            
        # Calculate recommendation scores
        scores = np.zeros(len(self.news_ids))
        valid_click_count = 0
        
        for news_id in user_clicked:
            if news_id in self.news_to_idx:
                row_index = self.news_to_idx[news_id]
                row = self.hybrid_similarity_matrix[row_index]
                if sp.issparse(row):
                    row_scores = row.toarray().flatten()
                else:
                    row_scores = np.asarray(row).flatten()
                scores += row_scores
                valid_click_count += 1
                
        if valid_click_count == 0:
            print(f"No valid clicked news found for user {user_id}")
            return []
            
        scores /= valid_click_count
        
        # Get recommendations
        clicked_set = set(user_clicked)
        sorted_indices = np.argsort(scores)[::-1]
        
        recommendations = []
        for idx in sorted_indices:
            if len(recommendations) >= top_n:
                break
            candidate = self.news_ids[idx]
            if candidate not in clicked_set:
                recommendations.append(self.get_news_details(candidate))
                
        return recommendations


    # Initialize recommender
recommender = NewsRecommender('updated.csv')
    
    # Example usage
user_id = 'U0'  # Replace with actual user ID
    
    # Get user's reading history
history = recommender.get_user_history(user_id)
print(f"\nUser {user_id}'s reading history:")
for news_id in history:
    news = recommender.get_news_details(news_id)
    print(f"\nTitle: {news['title']}")
    print(f"Category: {news['category']}")
        
    # Get recommendations
recommendations = recommender.recommend_news(user_id)
print(f"\nTop 5 recommendations for user {user_id}:")

```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import scipy.sparse as sp
from sklearn.model_selection import train_test_split
from sentence_transformers import SentenceTransformer  # For advanced text embeddings

class NewsRecommender:
    def __init__(self, data_path):
        """Initialize the recommender system with data path"""
        self.cleaned_data = pd.read_csv(data_path)
        self.prepare_data()
        self.create_similarity_matrices()
        
    def prepare_data(self):
        """Prepare and validate the data"""
        # Aggregate news content
        self.news_grouped = self.cleaned_data.groupby('News ID')['Combined'].agg(
            lambda texts: ' '.join(texts.astype(str))
        ).reset_index()
        
        # Add Main Category
        self.news_grouped = self.news_grouped.merge(
            self.cleaned_data[['News ID', 'Main Category']].drop_duplicates(),
            on='News ID',
            how='left'
        )
        
        # Create news ID mapping
        self.news_ids = self.news_grouped['News ID'].tolist()
        self.news_to_idx = {news_id: idx for idx, news_id in enumerate(self.news_ids)}
        
    def create_similarity_matrices(self):
        """Create content and category similarity matrices"""
        # Content-based similarity using advanced embeddings (e.g., Sentence Transformers)
        model = SentenceTransformer('all-MiniLM-L6-v2')
        self.content_embeddings = model.encode(self.news_grouped['Combined'])
        self.content_similarity_matrix = cosine_similarity(self.content_embeddings)
        
        # Category-based similarity
        categories = self.news_grouped['Main Category'].unique()
        category_to_idx = {cat: idx for idx, cat in enumerate(categories)}
        category_matrix = np.zeros((len(self.news_grouped), len(categories)))
        
        for idx, row in self.news_grouped.iterrows():
            category_matrix[idx, category_to_idx[row['Main Category']]] = 1
            
        self.category_similarity_matrix = cosine_similarity(category_matrix)
        
        # Create hybrid similarity matrix
        self.hybrid_similarity_matrix = 0.7 * self.content_similarity_matrix + 0.3 * self.category_similarity_matrix
        
    def get_user_history(self, user_id):
        """Get the news reading history for a user"""
        return self.cleaned_data[self.cleaned_data['User ID'] == user_id]['News ID'].unique()
        
    def get_news_details(self, news_id):
        """Get details of a specific news article"""
        news_info = self.cleaned_data[self.cleaned_data['News ID'] == news_id].iloc[0]
        return {
            'news_id': news_id,
            'title': news_info['News Title'],
            'content': news_info['News'],
            'category': news_info['Main Category']
        }
        
    def recommend_news(self, user_id=None, clicked_news_ids=None, top_n=5):
        """Generate news recommendations based on user's clicked news or user_id."""
        if clicked_news_ids is None:
            if user_id is None:
                raise ValueError("Either user_id or clicked_news_ids must be provided.")
            clicked_news_ids = self.get_user_history(user_id)
        
        if not clicked_news_ids.size:
            return []
        
        # Calculate recommendation scores
        scores = np.zeros(len(self.news_ids))
        valid_click_count = 0
        
        for news_id in clicked_news_ids:
            if news_id in self.news_to_idx:
                row_index = self.news_to_idx[news_id]
                row = self.hybrid_similarity_matrix[row_index]
                scores += row
                valid_click_count += 1
        
        if valid_click_count == 0:
            return []
        
        scores /= valid_click_count
        
        # Get recommendations
        clicked_set = set(clicked_news_ids)
        sorted_indices = np.argsort(scores)[::-1]
        
        recommendations = []
        for idx in sorted_indices:
            if len(recommendations) >= top_n:
                break
            candidate = self.news_ids[idx]
            if candidate not in clicked_set:
                recommendations.append(self.get_news_details(candidate))
        
        return recommendations
    
    def evaluate_recommendations(self, top_n=5, test_size=0.2, random_state=None):
        """Evaluate the recommender using Category Match and Content Similarity."""
        category_matches = []
        content_similarities = []
        
        # Iterate over all users
        for user_id in self.cleaned_data['User ID'].unique():
            user_clicks = self.get_user_history(user_id)
            if len(user_clicks) < 2:  # Skip users with insufficient history
                continue
            
            # Split clicks into train and test
            split_point = int(len(user_clicks) * (1 - test_size))
            train_clicks = user_clicks[:split_point]
            test_clicks = user_clicks[split_point:]
            
            if len(train_clicks) == 0 or len(test_clicks) == 0:
                continue
            
            # Generate recommendations based on train clicks
            recommendations = self.recommend_news(clicked_news_ids=train_clicks, top_n=top_n)
            if not recommendations:
                continue
            
            # Get user's main category
            user_main_category = self.cleaned_data[self.cleaned_data['User ID'] == user_id]['Main Category'].mode()[0]
            
            # Calculate category match
            category_match = len([news for news in recommendations if news['category'] == user_main_category]) / top_n
            category_matches.append(category_match)
            
            # Calculate content similarity
            user_content_embeddings = self.content_embeddings[[self.news_to_idx[news_id] for news_id in train_clicks if news_id in self.news_to_idx]]
            if len(user_content_embeddings) == 0:
                continue
            
            # Average content embedding for user's history
            user_avg_embedding = np.mean(user_content_embeddings, axis=0)
            
            # Content similarity for recommendations
            recommended_embeddings = self.content_embeddings[[self.news_to_idx[news['news_id']] for news in recommendations]]
            content_similarity = np.mean([cosine_similarity([user_avg_embedding], [embedding])[0][0] for embedding in recommended_embeddings])
            content_similarities.append(content_similarity)
        
        # Aggregate metrics
        avg_category_match = np.mean(category_matches) if category_matches else 0
        avg_content_similarity = np.mean(content_similarities) if content_similarities else 0
        
        return {
            f'CategoryMatch@{top_n}': avg_category_match,
            f'ContentSimilarity@{top_n}': avg_content_similarity
        }


# Initialize recommender
recommender = NewsRecommender('updated.csv')

# Evaluate the system
metrics = recommender.evaluate_recommendations(top_n=5, test_size=0.2, random_state=42)
print("\nEvaluation Metrics:")
for metric, value in metrics.items():
    print(f"{metric}: {value:.4f}")
```
